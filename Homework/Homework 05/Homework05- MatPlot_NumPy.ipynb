````python
x = np.linspace(0, 10, 100)  # X values from 0 to 10
y = np.sin(x)  # Y values as the sine of X
y2 = np.random.rand(10)  # Random Y values for scatter plot
labels = ['A', 'B', 'C', 'D', 'E']  # Labels for pie chart
sizes = [15, 30, 45, 10]  # Sizes for pie chart
area = np.pi * (15 * np.random.rand(10))**2  # Area for scatter plot circles
````

````python
# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='green')
plt.title('Line Plot of Sine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
````

````python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(x[:10], y2, s=area, alpha=0.5, color='red')
plt.title('Scatter Plot with Varying Sizes')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
````

````python
# Pie Plot
# Data for the pie chart
labels = ['A', 'B', 'C', 'D', 'E']  # Labels for each slice
sizes = [15, 30, 45, 10, 5]  # Sizes of each slice
colors = ['gold', 'red', 'lightskyblue', 'lightgreen', 'lightpink']  
explode = (0.1, 0, 0, 0, 0)  # Explode the 1st slice (A)

# Check if sizes and labels have the same length
if len(sizes) != len(labels):
    raise ValueError("The length of sizes must match the length of labels.")

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Pie Chart of Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

````python
# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.plot(x, y, color='Slateblue', alpha=0.6, linewidth=2)
plt.title('Area Plot of Sine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
````
